```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Best_Books_Ever_cleaned.xlsx")
df.head(5)
```

```python
columns_to_exclude = ["book_id", "description", "language", "isbn", "genres", "characters", "book_format", "edition", "pages", "number_of_ratings", "ratings_by_stars", "liked_percent", "setting", "cover_image", "bbe_score", "bbe_votes", "price","awards"]
df = df[df.columns.difference(columns_to_exclude)]
date_columns = ["first_publish_date", "publish_date"]
for column in date_columns:
    df[column] = pd.to_datetime(df[column], errors = "coerce")

df.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import altair as alt
```

```python
total_publishers = len(df["publisher"].unique())
print(total_publishers)
```

### In total, there are 2476 distinct publishing houses who have published award winning books. However, for the purposes of this project, we will only focus on the top 100 publishing houses shown below.

```python
total_awards_per_house = df.groupby("publisher")["number_of_awards"].sum()
total_awards_per_house = total_awards_per_house.sort_values(ascending=False)
total_awards_per_house.head(20)
```

```python
total_awards_per_house.head(20).plot(kind="bar")
plt.show()
```

```python
chart = alt.Chart(total_awards_per_house.head(20).reset_index()).mark_bar(color="teal").encode(
    x=alt.X('publisher:N', title='Publisher', sort=alt.EncodingSortField(field='number_of_awards', order='descending')),
    y=alt.Y('number_of_awards:Q', title='Total Number of Awards'),
    tooltip=['publisher:N', 'number_of_awards:Q']
).properties(
    width=600,
    title='Top 20 Publishers by Total Number of Awards'
)

chart
```

```python
top_4 = total_awards_per_house.head(4).index.tolist()

df_houses = df[df["publisher"].isin(top_4)]
df_houses.head(5)
```

```python
brush = alt.selection_interval()

points = alt.Chart(df_houses).mark_point().encode(
    x='rating:Q',
    y='number_of_awards:Q',
    color='publisher:N'
).add_params(
    brush
)

bars = alt.Chart(df_houses).mark_bar().encode(
    x='count()',
    y='publisher:N',
    color='publisher:N'
).transform_filter(
    brush
)

points & bars
```
